```python
#1. You are given two strings s and t. String t is generated by randomly shuffling string s and then adding one more 
#letter at a random position. Return the letter that was added to t
def find_the_difference(s,t):
    s_lst = list(s)
    t_lst = list(t)
    for elem in t_lst:
        if elem in s_lst:
            s_lst.pop(s_lst.index(elem))
        else:
            return elem
find_the_difference("ae", "aea") 
        
```

```python
#2. Given a function that accepts unlimited arguments, check and count how many data types 
# are in those arguments. Finally return the total in a list
def count_datatypes(*args):
    int_count=0
    str_count =0
    bool_count= 0
    list_count = 0
    tuple_count = 0
    dict_count = 0
    type_total_lst = []
    for i in args:
        if (type(i) is int):
            int_count +=1
        elif (type(i) is str):
            str_count +=1
        elif (type(i) is bool):
            bool_count +=1
        elif (type(i) is list):
            list_count +=1
        elif (type(i) is tuple):
            tuple_count +=1
        elif (type(i) is dict):
            dict_count +=1
    type_total_lst.append(int_count)
    type_total_lst.append(str_count)
    type_total_lst.append(bool_count)
    type_total_lst.append(list_count)
    type_total_lst.append(tuple_count)
    type_total_lst.append(dict_count)
    return type_total_lst
count_datatypes(4, 21, ("ES", "EN"), ("a", "b"), False, [1, 2, 3], [4, 5, 6])
```

```python
#3. A Fibonacci string is a precedence of the Fibonacci series. It works with any two characters of the English alphabet

def fib_str(num, lst_val):
    if (num ==2):
        return (', ').join(lst_val)
    for i in range(3, num+1):
        if(i==3):        
            new_lst_val = lst_val.copy()
            new_lst_val.sort()        
            lst_val.append(('').join(new_lst_val))
        else:
            lst_val.append(lst_val[i-2] + lst_val[i-3])
        
    return (', ').join(lst_val)
fib_str(6, ["n", "k"]) 
```

```python
#4. Given an integer between 0 and 26, make a variable (self.answer). That variable would be assigned to 
# a string in this format:

def ones_threes_nines(num):
    str_val = "nines:your answer, threes:your answer, ones:your answer"
    answer = 0
    
    while (num-9>=0):
        num -= 9
        answer+=1    
    str_val = str_val.replace("your answer", str(answer),1)
    
    answer = 0
    while (num-3>=0):
        num -= 3
        answer+=1    
    str_val = str_val.replace("your answer", str(answer),1)
    
    answer = 0
    while (num-1>=0):
        num -= 1
        answer+=1    
    str_val = str_val.replace("your answer", str(answer),1)
    
    return str_val
ones_threes_nines(10) 
```

```python
#5.Write a non-recursive function that takes an integer n and returns the Fibonacci sequence's value at index n.

def fib(n):
    fibs = [0, 1]
    for i in range(2, n+1):
        fibs.append(fibs[-1] + fibs[-2])        
    return fibs[n]

fib(6)

```
