```python
#1. Create a class Sudoku that takes a string as an argument. The string will contain the numbers of a regular 9x9 sudoku 
# board left to right and top to bottom, with zeros filling up the empty cells.
class Sudoku:
  def __init__(self, num_str_val):
    self.board= []
    tmp_lst_val = []
    for i in range(1, len(num_str_val)+1):
      tmp_lst_val.append(int(num_str_val[i-1]))
      if(i % 9 == 0):
        self.board.append(tmp_lst_val)
        tmp_lst_val = []
  
  def get_row(self, row_num):
    return self.board[row_num]
  
  def get_col(self, col_num):
    col = []
    for i in range(0, 8):
      col.append(self.board[i][col_num])
    return col
  def get_sqr(self, *args):
    sudoku_cell_dict = {}
    row_end = 0
    col_end = 0
    sqr_index = 0
    for i in range(0,9):
      tmp_lst_val = []
      if i%3 == 0:
        row_end += 3
        col_end = 3
      else:
        col_end += 3
      for row in range(row_end-3, row_end):
        for col in range(col_end-3, col_end):
            tmp_lst_val.append([row,col])
            if len(args) == 2:
              if (row == args[0] and col == args[1]):
                sqr_index = i

      sudoku_cell_dict[i] = tmp_lst_val
    sqr_lst = []
    if len(args) ==1:
      for index in sudoku_cell_dict[args[0]]:
        sqr_lst.append(self.board[index[0]][index[1]])
    else:
      for index in sudoku_cell_dict[sqr_index]:
        sqr_lst.append(self.board[index[0]][index[1]])

    return sqr_lst
game = Sudoku("417950030000000700060007000050009106800600000000003400900005000000430000200701580")
game.board      
game.get_row(0)
game.get_col(8)
game.get_sqr(1)
game.get_sqr(1, 8)
game.get_sqr(8, 3)
```

```python
#3. Write a class called CoffeeShop, which has three instance variables:
class CoffeeShop:
    def __init__(self,name, menu):
        self.name = name
        self.menu = menu
        self.orders = []
    
    def add_order(self, ordered_item): 
        for item in self.menu:
            if item["name"] == ordered_item:
                self.orders.append(ordered_item)
                return "Order added!"        
        return "This item is currently unavailable!"
        
    def fulfill_order(self): 
        if (len(self.orders) == 0):
            return "All orders have been fulfilled!"
        else:
            item = self.orders[0]
            self.orders.pop(0)
            return f"The {item} is ready!"
                    
    def list_orders(self): 
        return self.orders
    
    def due_amount(self): 
        total_amount = 0.0
        for ordered_item in self.orders:
            for item in self.menu:
                if item["name"] == ordered_item:
                    total_amount += item["price"]
        return total_amount
    
    def cheapest_item(self): 
        min_price = self.menu[0]["price"]
        menu_item = self.menu[0]["name"]
        
        for item in self.menu:
            if min_price > item["price"]:
                min_price = item["price"]
                menu_item = item["name"]
        return menu_item
    
    def drinks_only(self): 
        return [item["name"] for item in self.menu if item["type"] == "drink" ]
    
    def food_only(self):
        return [item["name"] for item in self.menu if item["type"] == "food" ]
        
menu = [
    {"name":"orange juice", "type":"drink", "price":1.50},
    {"name":"lemonade", "type":"drink", "price":0.75},
    {"name":"cranberry juice", "type":"drink", "price":1.75},
    {"name":"pineapple juice", "type":"drink", "price":2.00},
    {"name":"lemon iced tea", "type":"drink", "price":2.10},
    {"name":"vanilla chai latte", "type":"drink", "price":2.30},
    {"name":"hot chocolate", "type":"drink", "price":1.60},
    {"name":"iced coffee", "type":"drink", "price":1.17},
    {"name":"tuna sandwich", "type":"food", "price":3.00},
    {"name":"ham and cheese sandwich", "type":"food", "price":3.50},
    {"name":"bacon and egg", "type":"food", "price":3.75},
    {"name":"steak", "type":"food", "price":4.00},
    {"name":"hamburger", "type":"food", "price":4.50},
    {"name":"cinnamon roll", "type":"food", "price":1.00}   
]

tcs = CoffeeShop("Tesha",menu)
print(tcs.add_order("hot cocoa"))

print(tcs.add_order("iced tea"))


print(tcs.add_order("cinnamon roll"))
print(tcs.add_order("iced coffee"))
print(tcs.list_orders())
print(tcs.due_amount())

print(tcs.fulfill_order())
print(tcs.fulfill_order())
print(tcs.fulfill_order())


print(tcs.list_orders())


print(tcs.due_amount())


print(tcs.cheapest_item())
print(tcs.drinks_only())
print(tcs.food_only() )
```

```python
#4. In this challenge, write a funcion loneliest_number to find the last Lonely number inside a sequence. A number is Lonely 
#if the distance from its closest Prime sets a new record of the sequence.

def get_prime_list(lo,hi):
    prime_lst = []
    if lo < 2:
        lo = 2
    for num in range(lo,hi):
        prime = True
        for i in range(2,num):
            if (num%i==0):
                prime = False
        if prime:
            prime_lst.append(num)
    return prime_lst

def set_prime_distance(lonely_num_lst_val,num, distance, prime):
    if len(lonely_num_lst_val)==0 or (lonely_num_lst_val[len(lonely_num_lst_val)-1]["distance"] < distance):
        lonely_num_lst_val.append({"number": num, "distance":distance, "closest":prime})

def loneliest_number(lo, hi):
    lonely_num_lst_val =[]
    prime_lst = get_prime_list(lo,hi)
    for num in range(lo, hi+1):
        for j in range(0, len(prime_lst)):
            if j ==0:
                if num < prime_lst[j]:
                        distance = prime_lst[j] - num
                        set_prime_distance(lonely_num_lst_val,num, distance, prime_lst[j])
                        break
                elif num == prime_lst[j]:
                        distance = prime_lst[j+1] - num
                        set_prime_distance(lonely_num_lst_val,num, distance, prime_lst[j+1])
                        break
            elif j == len(prime_lst)-1:
                if num > prime_lst[j]:
                    distance =  num - prime_lst[j] 
                    set_prime_distance(lonely_num_lst_val,num, distance, prime_lst[j])
                    break
                elif num == prime_lst[j]:
                    distance =  num - prime_lst[j-1]
                    set_prime_distance(lonely_num_lst_val,num, distance, prime_lst[j-1])
                    break
            else:
                distance1 = num - prime_lst[j-1]
                distance2 = prime_lst[j+1] - num
                if distance1 >= distance2:
                    set_prime_distance(lonely_num_lst_val,num, distance2, prime_lst[j+1])
                    break
                elif distance1 < distance2:
                    set_prime_distance(lonely_num_lst_val,num, distance2, prime_lst[j-1])
                    break
    return lonely_num_lst_val
loneliest_number(8, 123)   
        
```

```python
#5. Implement a class Selfie that can store the current state of the object in the form of binary string. It can take multiple
# pictures and then recover to a state it was before. During testing an object will be provided with new attributes and
# their values. It will store its state. Then the values will be changed. Then it will be given new attributes. It will
# store its state again. It will be repeated few times.

import shelve
class Selfie:
    def __init__(self):
        self.state_index = 0
    
    def save_state(self):
        s = shelve.open('curr_obj_state.db')
        try:
            s[str(self.state_index)] = self
            self.state_index += 1
        finally:
            s.close()
    
    def recover_state(self, n):
        s = shelve.open('curr_obj_state.db')
        try:
            if (n < 0 or n >= len(s)):
                return self
            else:
                i = len(s)-1
                while(i> n):
                    del s[str(i)]
                    i -= 1                
                return s[str(n)]
        finally:
            s.close()
    
    def n_states(self):
        return self.state_index
    
p = Selfie()
p.x = 2
p.save_state()
p.x = 5
p = p.recover_state(0)
p.x
```
